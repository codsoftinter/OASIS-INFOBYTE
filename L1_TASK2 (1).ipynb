```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop(columns = ['Z_CostContact','Z_Revenue'])
```

```python
df.describe()
```

```python
df.mode().dropna()
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
df['In_relationship'] = df.apply(get_relationship,axis = 1)
df.head()
```

```python
df.rename(columns = {'education_2n Cycle\t' : 'education_2n Cycle'},inplace = True)
```

```python
def Education_status(row):
    if row['education_2n Cycle'] == 1:
        return '2'
    elif row['education_Basic'] ==1:
        return '1'
    elif row['education_Graduation'] == 1:
        return '3'
    elif row['education_Master'] == 1:
        return '4'
    elif row['education_PhD'] == 1:
        return '5'
    else:
        return 'unknown'
df['Education_status'] = df.apply(Education_status, axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_cluster = ['Income', 'MntTotal', 'In_relationship', 'Education_status']

df_scaled = df.copy()
df_scaled[cols_to_cluster] = scaler.fit_transform(df[cols_to_cluster])
```

```python
X = df_scaled[cols_to_cluster]
inertia_list = []
for k in range(1,11):
    inertia = KMeans(n_clusters=k, n_init=10).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(1,11),inertia_list)
plt.title("Inertia vs. Numbers of Cluster")
plt.xlabel('Numbers of Cluster (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters = 4, random_state = 7)
model.fit(df_scaled[cols_to_cluster])
df_scaled['Cluster'] = model.predict(df_scaled[cols_to_cluster])
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(df_scaled[cols_to_cluster])
df_scaled['pc1'] = pca_res[:,0]
df_scaled['pc2'] = pca_res[:,1]
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(x='pc1',y='pc2',data = df_scaled, hue = 'Cluster',palette = 'viridis')
plt.title('Cluster Visualization')
plt.xlabel('Princpal component 1 (pc1)')
plt.ylabel('Princpal component 2 (pc2)')
plt.legend(title = 'Clusters')
```

```python
customer_num = df_scaled.groupby('Cluster')['MntTotal'].count().reset_index()

plt.figure(figsize = (8,6))
sns.barplot(x='Cluster',y='MntTotal',data = customer_num,palette = 'viridis')
plt.title('Cluster sizes')
plt.xlabel('Cluster')
plt.ylabel('customer_num')
plt.show()
```

```python
scaler = StandardScaler()
cols_to_cluster = ['Income','MntTotal', 'In_relationship', 'Education_status']
df_Clustered = df.copy()

model = KMeans(n_clusters = 4, random_state = 7)
model.fit(df_Clustered[cols_to_cluster])
df_Clustered['Cluster'] = model.predict(df_Clustered[cols_to_cluster])
df_Clustered.head()
```

```python
df_Clustered['Avg_income'] = df_Clustered.groupby('Cluster')['Income'].mean()

plt.figure(figsize = (8,6))
sns.barplot(x='Cluster',y='Avg_income',data = df_Clustered ,palette = 'viridis')
plt.title('Cluster Avg_income')
plt.xlabel('Clusters')
plt.ylabel('Avg_income')
plt.show()
```

```python
Avg_consumption = df_Clustered.groupby('Cluster')[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','MntTotal']].mean().reset_index()
```

```python
Avg_consumption
```

```python
melted_data = pd.melt(Avg_consumption, id_vars="Cluster", var_name="Product", value_name="Consumption")
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x='Cluster',y='Consumption',hue = "Product",data = melted_data ,errorbar = None)
plt.title('Product Consumption by Cluster')
plt.xlabel('Clusters')
plt.ylabel('Product Consumption')
plt.xticks(rotation = 0)
plt.legend(title = "Product",loc = "upper right")

plt.show()
```
